# Zomato Data Analysis Project

## Importing Libraries

```python
import pandas as pd             # It is used for data manipulation and analysis.
import numpy as np              # It is used for numerical operations
import matplotlib.pyplot as plt # Use for data visualization
import seaborn as sns           # Use for data visualization
```

## Create the Data Frame
## Load CSV file

```python
df = pd.read_csv("zomato_data.csv")
print(df)
```

```python
df
```

```python
# DataFrame Info
df.info()
```

```python
# Check Missing Values
df.isnull().sum()
```

### Convert the data type of "rate" column: eg: 4.1/5 to 4.1

```python
df.rate
```

```python
def hanldeRate(value):                      # Create User defined function
    value = str(value).split('/')           # Pass the value
    value = value[0];
    return float(value)

df['rate'] = df['rate'].apply(hanldeRate)   #
print(df.head())
```

```python
# Find Data Types
df.dtypes
```

## Insights

### 1) Which type of restaurant is most popular among customers for placing orders?

#### Find Type of Restaurant

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['listed_in(type)'], hue=df['listed_in(type)'], palette='viridis', legend=False)
plt.xlabel('Type of Restaurant')
plt.ylabel('Order Count')
plt.title("Most Popular Restaurant Types for Orders")
#plt.xticks(rotation=45)
plt.show()
```

### 2) What is the total number of votes received by each restaurant type from customers?"

```python
df.head()
```

```python
# We need to group data by "type" and "vote" then sum it
group_data = df.groupby('listed_in(type)')['votes'].sum()
group_data
```

```python
#plt.figure(figsize=(8,6))
plt.plot(group_data, c='blue', marker='o')
plt.xlabel("Type of Restaurant", c='brown', size=15)
plt.ylabel('Total Votes', c='green', size=15)
plt.title('Total Votes Received by Each Restaurant Type', size=15)
plt.grid()
plt.show()
```

### 3) What are the most common ratings given to the majority of restaurants?

```python
plt.hist(df['rate'],bins=10, color='skyblue', edgecolor='black')
plt.title("Ratings Distribution of Restaurants", c='green',size=15)
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()         
```

### 4) What is the average amount couples spend on each online order on Zomato?

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x=couple_data, hue=couple_data, palette='magma')
plt.xlabel('Approximate Cost for Two People')
plt.ylabel('Count')
plt.title('Approximate Cost by Couple on Zomato Orders')
plt.xticks(rotation=45)
plt.show()
```

### 5) Which restaurant category receives the most offline orders, so that Zomato can target these customers with special offers?

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True,  cmap='YlGnBu',fmt='d')
plt.title("Online vs Offline Orders by Restaurant Category")
plt.xlabel("Online Order")
plt.ylabel("Restaurant Type")
plt.show()
```

### 6) Which ordering mode, online or offline, has received the highest customer ratings on Zomato?

```python
df.head(1)
```

```python
plt.figure(figsize=(7,7))
sns.boxplot(x='online_order', y='rate', data=df, palette='coolwarm', hue='online_order')
plt.xlabel("Online Order Available")
plt.ylabel("Restaurant Rating")
plt.title("Comparison of Ratings by Ordering Mode")
plt.show()
```

Ans: Dining restaurants get most of their orders in person or offline, while cafes get more orders online. This means people like to visit dining restaurants to order, but they prefer to order online from cafes.

### 7) Top 5 Most Popular Restaurants by Votes

```python
top_5_restaurants = df[['name','votes']].sort_values(by='votes', ascending=False).head(6)
#print('Top 5 Most Popular Restaurants by Votes: \n', top_5_restaurants)

sns.barplot(x='votes', y='name', data=top_5_restaurants, hue='votes', palette='viridis')
plt.title('Top 5 Most Popular Restaurants by Votes')
plt.xlabel("Votes")
plt.ylabel("Restaurant Name")
plt.show()
```

### 8) Average Rating by Restaurant Type:

```python
# Group "type" & "rate" and calculate the mean
avg_rating_by_type = df.groupby('listed_in(type)')['rate'].mean().sort_values(ascending=False)
print("Average Cost by Restaurants Type: \n", avg_rating_by_type)
```

### 9) Cost Analysis by Restaurant Type:

```python
# Group "type" & "approx_cost(for two people)"

avg_cost_by_type = df.groupby('listed_in(type)')['approx_cost(for two people)'].mean()
print("Average Cost by Restaurant Type:\n", avg_cost_by_type)
```

### 10) Count of Restaurants by Rating Range

```python
rating_bins = [0, 2, 3, 4, 4.5, 5]
df['rating_range'] = pd.cut(df['rate'], bins=rating_bins, labels=["0-2", "2-3", "3-4", "4-4.5", "4.5-5"])
rating_range_counts = df['rating_range'].value_counts().sort_index()
print("Count of Restaurants by Rating Range:\n", rating_range_counts)
```

### 11) Impact of Table Booking on Ratings

```python

plt.figure(figsize=(8,6))
sns.boxplot(x='book_table', y='rate', data=df, palette='Set2', hue='book_table')
plt.title("Impact of Table Booking on Ratings")
plt.xlabel("Table Booking Available")
plt.ylabel("Rating")
plt.show()
```

### 12) Correlation Between Votes and Ratings

```python
corr_votes_rating = df['votes'].corr(df['rate'])
print("Correlation Between Votes and Ratings:", corr_votes_rating)

sns.regplot(x='votes', y='rate', data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title("Correlation Between Votes and Ratings")
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.show()
```
